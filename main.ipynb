```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_excel('Ruta_Vendedor.xlsx')
data.head()
```

```python
data['Fecha'] = pd.to_datetime(data['Fecha'])
data['Dia_Semana'] = data['Fecha'].dt.day_name()

data['Rechazo'] = data['Total'] < 0

#Mostrar las siguientes columnas
data[['Fecha', 'Dia_Semana', 'Codigo_Cliente', 'Nombre_Cliente', 'Total', 'Rechazo']].head()
```

```python
data['Total_Compra'] = data.groupby(['Codigo_Cliente', 'Fecha'])['Total'].transform('sum')
```

```python
columnas_a_eliminar = ['Codigo_FV', 'FV', 'Codigo_Proveedor', 'Nombre_Proveedor', 'Marca', 'Linea', 'Codigo_Producto', 'Nombre_Producto',
                       'Cantidad', 'Factor', 'ValorVenta', 'Total', 'Categoria_Cliente', 'Tipo_Cliente', 'TipoDocIdentidad', 'DocIdentidad',
                       'NumeroTelefono', 'VendedorZonaCodigo']

data = data.drop(columns=columnas_a_eliminar)
```

```python
data = data.drop_duplicates(subset=['Codigo_Cliente', 'Fecha'])
```

# Construcción del modelo

```python
data_enconded = pd.get_dummies(data[['Dia_Semana']], drop_first=True)

data_enconded['Rechazo'] = data['Rechazo']

X = data_enconded.drop('Rechazo', axis=1)
y = data_enconded['Rechazo']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)


print("Classification Report:")
print(classification_report(y_test, y_pred))

print("confusion Matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
# Crear un dataframe con los clientes únicos y sus días sin rechazos
data_no_rechazos = data[data['Rechazo'] == False]

# Agrupar por cliente y día de la semana para encontrar los días sin rechazos
clientes_dias_no_rechazo = data_no_rechazos.groupby(['Codigo_Cliente', 'Nombre_Cliente'])['Dia_Semana'].apply(lambda x: list(x.unique())).reset_index()

# Renombrar las columnas
clientes_dias_no_rechazo.columns = ['Codigo_Cliente', 'Nombre_Cliente', 'Dias_Sin_Rechazo']

# Mostrar el dataframe resultante
clientes_dias_no_rechazo.head()
```

```python
clientes_dias_no_rechazo.to_excel('clientes_dias_sin_rechazo.xlsx', index=False)
```

```python
import pandas as pd

# Supongamos que 'data' es el dataframe original que cargamos al principio
# Y 'clientes_dias_no_rechazo' es el dataframe procesado

# Unir el dataframe original con el dataframe de días sin rechazo
data_completo = data.merge(clientes_dias_no_rechazo, on=['Codigo_Cliente', 'Nombre_Cliente'], how='left')

# Guardar el dataframe completo a un archivo Excel
output_path_completo = 'clientes_dias_sin_rechazo_completo.xlsx'
data_completo.to_excel(output_path_completo, index=False)

print(f"Archivo exportado a {output_path_completo}")
```
